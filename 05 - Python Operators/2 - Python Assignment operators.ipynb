# Python Assignment operators
___


Assignment operators as the name suggests assigns the value to the operand. 

The in place operators also known as shortcut operators that include +=, -=, /=, %=, //= and **= 

allows you to write code like num= num +10 more concisely as num +=3.

<br>

<tr>
    <th style="width:10%; text-align: center;">Operator</th>
    <th style="width:40%; text-align: center;">Description</th>

    <th style="width:20%; text-align: center;">Similar To</th>

<tr>
<td style="text-align: center;">=</td>
<td style="text-align: left;">Assign value of right side of expression to left side operand</b></td>
<td style="text-align: center;">x = 5</td>
<td style="text-align: center;">x = 5</td>
</tr>
    
<tr>
<td style="text-align: center;">+=</td>
<td style="text-align: left;"><b>Add and Assign </b>: Add right side operand with left side operand and then assign to left operand</td>
<td style="text-align: center;">x += 5</td>
<td style="text-align: center;">x + 5</td>
</tr>

<tr>
<td style="text-align: center;">-=</td>
<td style="text-align: left;"><b>Subtract AND </b>: Subtract right operand from left operand and then assign to left operand: True if both operands are equal</td>
<td style="text-align: center;">x -= 5</td>
<td style="text-align: center;">x - 5</td>
</tr>

<tr>
<td style="text-align: center;">*=</td>
<td style="text-align: left;"><b>Multiply AND </b>: Multiply right operand with left operand and then assign to left operand</td>
<td style="text-align: center;">x *= 5</td>
<td style="text-align: center;">x * 5</td>
</tr>

<tr>
<td style="text-align: center;">/=</td>
<td style="text-align: left;"><b>Divide AND </b>: Divide left operand with right operand and then assign to left operand</td>
<td style="text-align: center;">x /= 5</td>
<td style="text-align: center;">x / 5</td>
</tr>
    
<tr>
<td style="text-align: center;">%=</td>
<td style="text-align: left;"><b>Modulus AND </b>: Takes modulus using left and right operands and assign result to left operand</td> 
<td style="text-align: center;">x %= 5</td>
<td style="text-align: center;">x % 5</td>
</tr>
    
<tr>
<td style="text-align: center;">//=</td>
    <td style="text-align: left;"><b>Divide(floor) AND </b>: Divide left operand with right operand and then assign the value(floor) to left operand</td>
<td style="text-align: center;">x //= 5</td>
<td style="text-align: center;">x // 5</td>
</tr>
    
<tr>
<td style="text-align: center;">**=</td>
<td style="text-align: left;"><b>Exponent AND </b>: Calculate exponent(raise power) value using operands and assign value to left operand</td>
<td style="text-align: center;">x **= 5</td>
<td style="text-align: center;">x ** 5</td>
</tr>
    
<tr>
<td style="text-align: center;">&=</td>
    <td style="text-align: left;">Performs Bitwise AND on operands and assign value to left operand</td>
<td style="text-align: center;">x &= 5</td>
<td style="text-align: center;">x & 5</td>
</tr>
    
<tr>
<td style="text-align: center;">|=</td>
    <td style="text-align: left;">Performs Bitwise OR on operands and assign value to left operand</td>
<td style="text-align: center;">x |= 5</td>
<td style="text-align: center;">x | 5</td>
</tr>
    
<tr>
<td style="text-align: center;">^=</td>
    <td style="text-align: left;">Performs Bitwise xOR on operands and assign value to left operand</td>
<td style="text-align: center;">x ^= 5</td>
<td style="text-align: center;">x ^ 5</td>
</tr>
    
<tr>
<td style="text-align: center;">>>=</td>
    <td style="text-align: left;">Performs Bitwise right shift on operands and assign value to left operand</td>
<td style="text-align: center;">x >>= 5</td>
<td style="text-align: center;">x >> 5</td>
</tr>
    
<tr>
<td style="text-align: center;"><<=</td>
    <td style="text-align: left;">Performs Bitwise left shift on operands and assign value to left operand</td>
<td style="text-align: center;">x <<= 5</td>
<td style="text-align: center;">x << 5</td>
</tr>
