# Which Ministry has the Highest Budget Allocation

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in â‚¹crores)',figsize=(10,5))
```

```python
data=data.iloc[:-1 , :]
```

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in â‚¹crores)',figsize=(18,8))
plt.title("Distribution of The Budget For Top 10 Ministries", fontweight='bold')
plt.show
```

# Distribution of The Budget For Top 10 Ministries

```python
data = data.sort_values(by='Fund allotted(in â‚¹crores)',ascending=False)
```

```python
new_data=data.head(10)
df = new_data["Fund allotted(in â‚¹crores)"]
labels = new_data["Department /Ministry"]
plt.figure(figsize=(10,9))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=45, pctdistance=0.7, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget For Top 10 Ministries", fontweight='bold')
plt.show()
```

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in â‚¹crores)',figsize=(18,8)) #you can use 'new_data' to see the only the top 10
plt.title("Distribution of The Budget For Top 10 Ministries", fontweight='bold')
plt.show
```

# Distribution of The Budget For Last 10 Ministries

```python
data = data.sort_values(by='Fund allotted(in â‚¹crores)',ascending=True)
new_data = data.head(10)
new_data
```

```python
df = new_data["Fund allotted(in â‚¹crores)"]
labels = new_data["Department /Ministry"]
plt.figure(figsize=(10,9))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=0, pctdistance=0.7, shadow =True)
central_circle = plt.Circle((0, 0), 0.8, color='white') #change the 0.8 to see changes to the circle
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget For Last 10 Ministries", fontweight='bold')
plt.show()
```

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in â‚¹crores)',figsize=(18,8))
plt.title("Distribution of The Budget For Last 10 Ministries", fontweight='bold')
plt.show
```
